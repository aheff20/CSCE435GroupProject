```python
gbs = {}

for i in tks.keys():
    if "mpi" in i:
        gbs[i + "_strong"] = tks[i].groupby("InputSize")
        gbs[i + "_weak"] = tks[i].groupby("values_per_proc")
    else:
        gbs[i + "_strong"] = tks[i].groupby("InputSize")
        gbs[i + "_weak"] = tks[i].groupby("num_blocks")
```

Compose these 3 groups column-wise to get a holistic view of the data

```python
print(gbs.keys())
```

```python
gbs["merge_mpi_sorted_weak"]
```

```python
ctks = {}

for i in gbs.keys():
    if "mpi" in i:
        ctks[i] = th.Thicket.concat_thickets(
            thickets=list(gbs[i].values()),
            headers=list(gbs[i].keys()),
            axis="columns",
            metadata_key="num_procs"
        )
    else:
        ctks[i] = th.Thicket.concat_thickets(
            thickets=list(gbs[i].values()),
            headers=list(gbs[i].keys()),
            axis="columns",
            metadata_key="num_threads"
        )
```

```python
ctks.keys()
```

```python
valsPerProcList = {}

valsPerProcList["sample"] = ctks["sample_mpi_random_weak"].dataframe.columns.get_level_values(0).unique().drop("name")
valsPerProcList["bubble"] = ctks["bubble_mpi_random_weak"].dataframe.columns.get_level_values(0).unique().drop("name")
valsPerProcList["bitonic"] = ctks["bitonic_mpi_random_weak"].dataframe.columns.get_level_values(0).unique().drop("name")
valsPerProcList["merge"] = ctks["merge_mpi_random_weak"].dataframe.columns.get_level_values(0).unique().drop("name")
```

```python
valsPerBlockList = {}

valsPerBlockList["sample"] = ctks["sample_cuda_random_weak"].dataframe.columns.get_level_values(0).unique().drop("name")
valsPerBlockList["bubble"] = ctks["bubble_cuda_random_weak"].dataframe.columns.get_level_values(0).unique().drop("name")
valsPerBlockList["bitonic"] = ctks["bitonic_cuda_random_weak"].dataframe.columns.get_level_values(0).unique().drop("name")
valsPerBlockList["merge"] = ctks["merge_cuda_random_weak"].dataframe.columns.get_level_values(0).unique().drop("name")
```

```python
mpiInputSizeList = {}

mpiInputSizeList["sample"] = ctks["sample_mpi_random_strong"].dataframe.columns.get_level_values(0).unique().drop("name")
mpiInputSizeList["bubble"] = ctks["bubble_mpi_random_strong"].dataframe.columns.get_level_values(0).unique().drop("name")
mpiInputSizeList["bitonic"] = ctks["bitonic_mpi_random_strong"].dataframe.columns.get_level_values(0).unique().drop("name")
mpiInputSizeList["merge"] = ctks["merge_mpi_random_strong"].dataframe.columns.get_level_values(0).unique().drop("name")
```

```python
cudaInputSizeList = {}

cudaInputSizeList["sample"] = ctks["sample_cuda_random_strong"].dataframe.columns.get_level_values(0).unique().drop("name")
cudaInputSizeList["bubble"] = ctks["bubble_cuda_random_strong"].dataframe.columns.get_level_values(0).unique().drop("name")
cudaInputSizeList["bitonic"] = ctks["bitonic_cuda_random_strong"].dataframe.columns.get_level_values(0).unique().drop("name")
cudaInputSizeList["merge"] = ctks["merge_cuda_random_strong"].dataframe.columns.get_level_values(0).unique().drop("name")
```

```python
valsPerProcList["merge"]
```

Perform pandas operations to get the table in a desirable format for plotting

```python
for i in ctks.keys():
    if "cuda" in i:
        if "strong" in i:
            for inputSize in cudaInputSizeList[i.split("_")[0]]:
                ctks[i].dataframe = ctks[i].dataframe.drop([(inputSize, "nid"), (inputSize, "spot.channel")], axis=1)
            ctks[i].dataframe = ctks[i].dataframe.reset_index().drop([("node", "")], axis=1)
            ctks[i].dataframe = ctks[i].dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])

        else:
            for valPerBlock in valsPerBlockList[i.split("_")[0]]:
                ctks[i].dataframe = ctks[i].dataframe.drop([(valPerBlock, "nid"), (valPerBlock, "spot.channel")], axis=1)
            ctks[i].dataframe = ctks[i].dataframe.reset_index().drop([("node", "")], axis=1)
            ctks[i].dataframe = ctks[i].dataframe.rename({("name", ""): "name", ("num_threads", ""): "num_threads"}).set_index(["name", "num_threads"])

    else:
        if "strong" in i:
            for inputSize in mpiInputSizeList[i.split("_")[0]]:
                ctks[i].dataframe = ctks[i].dataframe.drop([(inputSize, "nid"), (inputSize, "spot.channel")], axis=1)

            ctks[i].dataframe = ctks[i].dataframe.reset_index().drop([("node", "")], axis=1)
            ctks[i].dataframe = ctks[i].dataframe.rename({("name", ""): "name", ("num_procs", ""): "num_procs"}).set_index(["name", "num_procs"])

        else: 
            for valPerProc in valsPerProcList[i.split("_")[0]]:
                if(valPerProc == 512):
                    continue
                ctks[i].dataframe = ctks[i].dataframe.drop([(valPerProc, "nid"), (valPerProc, "spot.channel")], axis=1)

            ctks[i].dataframe = ctks[i].dataframe.reset_index().drop([("node", "")], axis=1)
            ctks[i].dataframe = ctks[i].dataframe.rename({("name", ""): "name", ("num_procs", ""): "num_procs"}).set_index(["name", "num_procs"])
            
```

# Plotting

### Option 1
Use `pandas.DataFrame.plot` to automatically interface with matplotlib

```python
def plot_pandas(df, title, xlabel, ylabel, loglog):
    plot = df.plot(kind="line", 
        marker='o', 
        title=title, 
        xlabel=xlabel,
        ylabel=ylabel, 
        grid=True,
        loglog=loglog
    )
    return plot
```

# MPI STRONG SCALING TIME PLOTS

```python
for name in function_names:
    for k in ["random", "reverse", "sorted", "perturbed"]:
        masterDF = pd.DataFrame()
        for inputType in ctks.keys():
            if "strong" in inputType and "mpi" in inputType and k in inputType:
                tempDF = ctks[inputType].dataframe.loc[name][(67108864, "Avg time/rank")]
                tempDF = tempDF.rename(inputType.split('_')[0])    
                masterDF = pd.concat([masterDF, tempDF], axis=1) 
        
        plot_pandas(
            df=masterDF,
            title="MPI, " + str(name) + ", Sorting " + str(67108864) + " Values, " + str(k) + " Input, Avg time/rank",
            xlabel="Num_Procs",
            ylabel="Time (seconds)",
            loglog=True
        )
```

# MPI WEAK SCALING TIME PLOTS

```python
for name in function_names:
    for k in ["random", "reverse", "sorted", "perturbed"]:
        masterDF = pd.DataFrame()
        for inputType in ctks.keys():
            if "weak" in inputType and "mpi" in inputType and k in inputType:
                tempDF = ctks[inputType].dataframe.loc[name][(1048576, "Avg time/rank")]
                tempDF = tempDF.rename(inputType.split('_')[0])    
                masterDF = pd.concat([masterDF, tempDF], axis=1) 
            
        masterDF = masterDF.apply(lambda x: x.fillna((x.shift() + x.shift(-1)) / 2))
            
        plot_pandas(
            df=masterDF,
            title="MPI, " + str(k) + " Input, " + str(name) + ", 1048576 values per proc, Avg time/rank",
            xlabel="Num Processors",
            ylabel="Time (seconds)",
            loglog=True
        )
            
```

## Speedup

```python

for name in function_names:
    for k in ["random", "reverse", "sorted", "perturbed"]:
        masterDF = pd.DataFrame()
        for inputType in ctks.keys():
            if "strong" in inputType and "mpi" in inputType and k in inputType:
                tempDF = ctks[inputType].dataframe.loc[name][(67108864, "Avg time/rank")]
                tempDF = tempDF.rename(inputType.split('_')[0])    
                masterDF = pd.concat([masterDF, tempDF], axis=1) 
                
        for alg in ["sample", "merge", "bubble", "bitonic"]:
            masterDF[alg] = masterDF[alg][2] / masterDF[alg]
        
        plot_pandas(
            df=masterDF,
            title="MPI Speed-up, " + str(name) + ", Sorting " + str(67108864) + " Values, " + str(k) + " Input",
            xlabel="Num_Procs",
            ylabel="Time (seconds)",
            loglog=True
        )
```

```python
for name in function_names:
    for k in ["random", "reverse", "sorted", "perturbed"]:
        masterDF = pd.DataFrame()
        for inputType in ctks.keys():
            if "strong" in inputType and "cuda" in inputType and k in inputType:
                tempDF = ctks[inputType].dataframe.loc[name][(1048576, "Avg time/rank")]
                tempDF = tempDF.rename(inputType.split('_')[0])    
                masterDF = pd.concat([masterDF, tempDF], axis=1) 
                
        for alg in ["sample", "merge", "bubble", "bitonic"]:
            masterDF[alg] = masterDF[alg][64] / masterDF[alg]
        
        plot_pandas(
            df=masterDF,
            title="CUDA Speed-up, " + str(name) + ", Sorting " + str(1048576) + " Values, " + str(k) + " Input",
            xlabel="Num_Blocks",
            ylabel="Time (seconds)",
            loglog=True
        )
```
